```python
codename = {}
for areacode in storeList[['시군구코드','시군구명']].values:
    if areacode[0] in codename:
        continue
    else:
        codename[areacode[0]]=areacode[1]
```

```python

choropleth = folium.Choropleth(geo_data = geo,
                 name = 'choropleth',
                 data = 유동인구순위, #   geo_data에 바인딩할 데이터, geo_data와 data의 개수가 일치해야 함
                 columns = ['자치구코드','첫번째'], # 2개 컬럼만 지정해야 함.
                 key_on = 'feature.properties.SIG_KOR_NM', # columns에 지정된 첫번째 컬럼과 매칭할 geo_data의 키값
                legend_name = '유동인구수',
                  fill_color= 'BuPu',
                  highlight = True
                 ).add_to(map3)
```

```python
#허버링(마우스오버) 시 구이름 나타나기
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['SIG_KOR_NM'])
)
```
